```python
from binance.client import Client

# Initialize the Binance client
api_key = 'your_api_key'    
api_secret = 'your_api_secret'
client = Client(api_key, api_secret)

# Fetch current price of a symbol
symbol = 'BTCUSDT'
current_price = client.get_symbol_ticker(symbol=symbol)
print(f"Current price of {symbol}: {current_price['price']}")

# Fetch 24-hour ticker price change statistics
ticker_24hr = client.get_ticker(symbol=symbol)
print(f"24-hour ticker price change statistics for {symbol}: {ticker_24hr}")

# Fetch order book
order_book = client.get_order_book(symbol=symbol)
print(f"Order book for {symbol}: {order_book}")

# Fetch recent trades
recent_trades = client.get_recent_trades(symbol=symbol)
print(f"Recent trades for {symbol}: {recent_trades}")
```

Relevant Data to Fetch
Current Price: Useful for checking the latest trading price of a cryptocurrency.
24-Hour Ticker Price Change Statistics: Provides information on price changes, volume, and other statistics over the past 24 hours.
Order Book: Shows the current buy and sell orders, which can be useful for understanding market depth and liquidity.
Recent Trades: Lists the most recent trades, which can help in analyzing market activity and trends.
